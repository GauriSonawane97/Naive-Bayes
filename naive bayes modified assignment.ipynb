```python
from sklearn import preprocessing
```

```python
salary_train['workclass'] = preprocessing.LabelEncoder().fit_transform(salary_train['workclass']) 
salary_train['education'] = preprocessing.LabelEncoder().fit_transform(salary_train['education']) 
salary_train['maritalstatus'] = preprocessing.LabelEncoder().fit_transform(salary_train['maritalstatus']) 
salary_train['occupation'] = preprocessing.LabelEncoder().fit_transform(salary_train['occupation']) 
salary_train['relationship'] = preprocessing.LabelEncoder().fit_transform(salary_train['relationship']) 
salary_train['race'] = preprocessing.LabelEncoder().fit_transform(salary_train['race']) 
salary_train['sex'] = preprocessing.LabelEncoder().fit_transform(salary_train['sex']) 
salary_train['native'] = preprocessing.LabelEncoder().fit_transform(salary_train['native']) 
```

```python
train_x = salary_train[salary_train.columns[0:13]]
train_y = salary_train[salary_train.columns[13]]
test_x = salary_test[salary_test.columns[0:13]]
test_y = salary_test[salary_test.columns[13]]
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
pred_gnb = GaussianNB().fit(train_x, train_y).predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(test_y, pred_gnb)
```

```python
print("Accuracy", (10759+1209)/(10759+1209+601+2491))
```

```python
pred_nmb = MultinomialNB().fit(train_x, train_y).predict(test_x)
```

```python
confusion_matrix(test_y, pred_nmb)
```
